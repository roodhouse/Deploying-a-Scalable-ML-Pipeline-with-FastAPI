## 1. Imports and load

```python
from pathlib import Path

import numpy as np
import pandas as pd

# Notebook lives in data/, so census.csv is next to this file
DATA_PATH = Path("census.csv")
if not DATA_PATH.exists():
    # Fallback if the kernel cwd is the project root
    DATA_PATH = Path("data/census.csv")

print("Loading:", DATA_PATH.resolve())
df = pd.read_csv(DATA_PATH)
print("Loaded successfully.")
```

## 6. Categorical overview (object columns)

```python
cat_cols = df.select_dtypes(include=["object", "string"]).columns.tolist()
print("Categorical columns:", cat_cols)

for col in cat_cols:
    nunique = df[col].nunique(dropna=False)
    print(f"\n=== {col} ({nunique} unique) ===")
    print(df[col].value_counts(dropna=False).head(15))
```

## 7. Target distribution (`salary`)

```python
if "salary" in df.columns:
    print(df["salary"].value_counts())
    print("\nProportions:")
    print(df["salary"].value_counts(normalize=True).round(4))
else:
    print("No 'salary' column found. Columns:", list(df.columns))
```

## 8. Check for messy spaces (common issue in this dataset)

The course notes that raw census data often has **leading spaces** in values or column names.  
These cells flag that so you know whether cleaning is needed.


```python
# Column names: leading/trailing whitespace?
col_issues = [(c, repr(c)) for c in df.columns if c != c.strip()]
print("Columns with leading/trailing spaces:", col_issues if col_issues else "None")

# String cells: any value that changes when stripped?
space_report = []
for col in df.select_dtypes(include=["object"]).columns:
    s = df[col].astype(str)
    # values that are not equal to stripped version
    bad = (s != s.str.strip()).sum()
    if bad:
        space_report.append((col, int(bad)))

print("\nObject columns with unstripped whitespace in cells:")
if space_report:
    for col, n in space_report:
        print(f"  {col}: {n} rows")
        # show a few examples
        examples = df.loc[df[col].astype(str) != df[col].astype(str).str.strip(), col].head(5).tolist()
        print("    examples:", examples)
else:
    print("  None detected (this copy may already be cleaned).")
```

```python
# Peek at raw file header (first 2 lines) for hidden spaces
raw_path = DATA_PATH
with open(raw_path, encoding="utf-8") as f:
    for i in range(2):
        line = f.readline()
        print(f"Line {i}: {line!r}")
```
